```python
#Visualização das colunas para seleção
dadosInse.columns.values
```

```python
#Seleção de colunas que seerão utilizadas
colSelec = ['SIGLA_UF', 'TP_LOCALIZACAO', 'QTD_ALUNOS_INSE', 'PC_NIVEL_1', 'PC_NIVEL_5', 'PC_NIVEL_8']
```

```python
#Novamente realizando a atribuição do novo filtro a uma variável
dadosInseUFLocalNivel = dadosInse.filter(items=colSelec)
```

```python
#Realizando a limpeza das células vazias
dadosInseUFLocalNivel = dadosInseUFLocalNivel.dropna()
```

```python
#Visualização para validação
dadosInseUFLocalNivel.head()
```

```python
#Criação de dicionário para interpretação dos dados contidos nas células
localDicion = {0:'Total Urbana e Rural',
1:'Urbana',
2:'Rural'}
```

```python
#Lista de fitros agrupados pela localização - Possível observar que o maior percuntual de alunos classificados no nível 1 estão licalizados na área Rural, sendo também a única área que tem o percentual de nível 8 menor que o de nível 1
dadosInseSelec.filter(items=['TP_LOCALIZACAO', 'PC_NIVEL_1', 'PC_NIVEL_5', 'PC_NIVEL_8']).groupby('TP_LOCALIZACAO').sum()
```

```python
#Teste de visualização do dicionário
localDicion[1]
```

```python
#Visualização da localização contida nas linhas
dadosInseSelec.TP_LOCALIZACAO
```

```python
#Criação de uma nova coluna para informar a mensagem do dicionário de acordo com o código apresentado na coluna TP_LOCALIZACAO
dadosInseSelec['NO_TP_LOCALIZACAO'] = [localDicion[resp] for resp in dadosInseSelec.TP_LOCALIZACAO]
```

```python
#Validação
dadosInseSelec.head()
```

```python
#Novo dicionário referente ao código do Tipo de Capital
capDicion = {0:'Total Capital e Interior',
1:'Capital',
2:'Interior'}
```

```python
#Criação de uma nova coluna para informar a mensagem do dicionário de acordo com o código apresentado na coluna TP_CAPITAL
dadosInseSelec['NO_TP_CAPITAL'] = [capDicion[resp] for resp in dadosInseSelec.TP_CAPITAL]
```

```python
#Validação
dadosInseSelec.head()
```

```python
#Visualização melhorada agupada pela localização apresentando os níveis selecionados organidas a partir da ordem decresente da coluna PC_NIVEL_1
dadosInseSelec.filter(items=['PC_NIVEL_1', 'PC_NIVEL_5', 'PC_NIVEL_8', 'NO_TP_LOCALIZACAO']).groupby('NO_TP_LOCALIZACAO').sum().sort_values(by='PC_NIVEL_1', ascending=False)
```

```python
#Visualização agrupada pelo Tipo de Capital - Como esperado, o mesmo acontece em relação ao tipo de capital, onde o Interior apresenta o maior percentual nível 1 e meor percentual nível 8
dadosInseSelec.filter(items=['PC_NIVEL_1', 'PC_NIVEL_5', 'PC_NIVEL_8', 'NO_TP_CAPITAL']).groupby('NO_TP_CAPITAL').sum().sort_values(by='PC_NIVEL_1', ascending=False)
```

```python
#Armazenado a informação em uma variável
localNivel = dadosInseSelec.filter(items=['PC_NIVEL_1', 'NO_TP_LOCALIZACAO']).groupby('NO_TP_LOCALIZACAO').sum().sort_values(by='PC_NIVEL_1', ascending=False)
```

```python
#Comando "plot()" para apresentar os dados verificados acima, referente a localização, de forma gráfica
localNivel.plot()
```

```python
#Atribuindo as informações do agrupamento por Tipo de Capital a uma variável
capNivel = dadosInseSelec.filter(items=['PC_NIVEL_1', 'NO_TP_CAPITAL']).groupby('NO_TP_CAPITAL').sum().sort_values(by='PC_NIVEL_1', ascending=False)
```

```python
#Comando "plot()" para apresentar os dados verificados acima, referente ao tipo de capital, de forma gráfica
capNivel.plot()
```

```python
#Incluindo coluna da variável principal a variável com seleção
dadosInseSelec['SIGLA_UF'] = dadosInse.SIGLA_UF
```

```python
#Validação
dadosInseSelec.head()
```

```python
#Utilização do "where()" para verificar se as informações obtidas acima permanecem ao utilizar apenas um estado filtrando as células com UF SP e apresentando o resultado
dadosInseSelec.filter(items=['PC_NIVEL_1', 'PC_NIVEL_5', 'PC_NIVEL_8', 'NO_TP_CAPITAL'])\
.where(dadosInse.SIGLA_UF == 'SP')\
.groupby('NO_TP_CAPITAL')\
.sum().sort_values(by='PC_NIVEL_1', ascending=False)
```

```python
#Utilizando o mesmo principio acima em um estado de condisões divergente para verificar a variação
dadosInseSelec.filter(items=['PC_NIVEL_1', 'PC_NIVEL_5', 'PC_NIVEL_8', 'NO_TP_CAPITAL'])\
.where(dadosInse.SIGLA_UF == 'MA')\
.groupby('NO_TP_CAPITAL')\
.sum().sort_values(by='PC_NIVEL_1', ascending=False)
```

```python
#Agrupando dados de Tipo de Capital por UF medindo os níveis 1, 5 e 8.
dadosInseSelec.filter(items=['SIGLA_UF', 'PC_NIVEL_1', 'PC_NIVEL_5', 'PC_NIVEL_8', 'NO_TP_CAPITAL'])\
.groupby(['SIGLA_UF', 'NO_TP_CAPITAL'])\
.sum()
```

```python
#Ajustes de título e tamanho do gráfico
fig, ax = plt.subplots(figsize=(20,7))
plt.suptitle('Nível Socioeconômico x Tipo de Capital x UF', fontsize=20)

#Apresentação de gráfico demostrando a comparação do Nível Socioeconômico em relação ao Tipo de Capital do aluno por UFs
dadosInseSelec.filter(items=['SIGLA_UF', 'PC_NIVEL_1', 'NO_TP_CAPITAL'])\
.groupby(['SIGLA_UF', 'NO_TP_CAPITAL'])\
.sum()\
.unstack().plot(ax=ax, )
```

```python
#Apresentação de gráfico demostrando a comparação do Nível Socioeconômico (Níveis 1 e 8) em relação ao Tipo de Capital
ax = dadosInseSelec.filter(items=['PC_NIVEL_1', 'PC_NIVEL_8', 'NO_TP_CAPITAL'])\
.groupby('NO_TP_CAPITAL')\
.sum()\
.plot(figsize=(20,7))
ax.set_title('Nível Socioeconômico x Tipo de Capital', fontsize=20)
```
